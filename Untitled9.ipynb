```python
import yfinance as yf
gold=yf.download("GC=F",period="1y")
print(gold.head())
```

```python
gold['MA20']=gold[('Close','GC=F')].rolling(20).mean()
gold['MA50']=gold[('Close','GC=F')].rolling(50).mean()
gold['Signal']=0
gold.loc[gold['MA20']>gold['MA50'],'Signal']=1
gold.loc[gold['MA20']<gold['MA50'],'Signal']=-1
gold.loc[gold['MA20']==gold['MA50'],'Signal']=0
gold['position']=gold['Signal'].diff()
gold[['Signal','position']].tail(10)
```

```python
gold[gold['position'] != 0][['Signal', 'position']].head(20)
```

```python
buy_signal = gold[gold['position'] > 0]
sell_signal = gold[gold['position'] < 0]

print("Buy Signals")
print(buy_signal[['Signal','position']])

print("Sell Signals")
print(sell_signal[['Signal','position']])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(gold[('Close','GC=F')],label='Gold price', alpha=0.5)
plt.plot(gold['MA20'],label='MA20', linestyle='--')
plt.plot(gold['MA50'],label='MA50', linestyle='--')

# Plotting all buy and sell signals found in the filtered dataframes
plt.scatter(buy_signal.index, buy_signal[('Close','GC=F')], marker='^', s=100, color='green', label='BUY', zorder=5)
plt.scatter(sell_signal.index, sell_signal[('Close','GC=F')], marker='v', s=100, color='red', label='SELL', zorder=5)

plt.legend()
plt.title(' DIGZZ GOLD TRADING STRATEGY ')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True, alpha=0.2)
plt.show()
```

```python
buy_price = gold.loc['2025-08-25',('Close','GC=F')]
sell_price= gold.loc['2026-03-26',('Close','GC=F')]
profit = sell_price - buy_price
print(f"Buy Price: {buy_price}")
print(f"Sell Price: {sell_price}")
print(f"Profit: {profit}")
```

```python
return_percentage = (sell_price - buy_price) / buy_price * 100
print(f"Return Percentage: {return_percentage}%")
```

```python
gold['Daily Return'] = gold[('Close','GC=F')].pct_change()
gold['Strategy Return'] = gold['Daily Return'] * gold['Signal'].shift(1)
print(gold[['Daily Return', 'Strategy Return']].tail())
```

```python
gold['Cumulative Strategy']=(1+gold['Strategy Return']).cumprod()
print(gold[['Cumulative Strategy']].tail())
```

```python
import matplotlib.pyplot as plt
gold['Cumulative Strategy'].plot(figsize=(12,6))
plt.title("Diganths Growth")
plt.xlabel("Date")
plt.ylabel("portfolio value")
plt.grid(True,alpha=0.3)
plt.show()
```
